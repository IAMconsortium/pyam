# Customizing Legends

This is a short tutorial showing a few examples of how different arguments to the `legend` keyword in various plotting methods affects where the legend is located.

First, let's get an example dataset.

```python
import pyam

df = (
    pyam.IamDataFrame(data='tutorial_AR5_data.csv', encoding='utf-8')
    .filter(variable='Emissions|CO2', region='World')
)

df.head()
```

By default, a legend will not appear if there are too many entries.

```python
df.line_plot()
```

By using the `color` argument, we restrict the number of legend entries, and it will show up with `matplotib` defaults.

```python
df.line_plot(color='model')
```

You can use standard `matplotlib` legend arguments by passing a dictionary of keyword arguments.

```python
df.line_plot(color='model', legend=dict(loc='center left'))
```

However, we also offer support for a few standard special cases, namely legends to the right of a plot and legends below the plot

```python
from pyam.plotting import OUTSIDE_LEGEND
```

```python
df.line_plot(color='model', legend=OUTSIDE_LEGEND['right'])
```

```python
df.line_plot(color='model', legend=OUTSIDE_LEGEND['bottom'])
```
